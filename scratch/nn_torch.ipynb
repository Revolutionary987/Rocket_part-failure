```python
import numpy as np

class Layer:
    def __init__(self,input_dim,output_dim):
        # Xavier initialisation

        self.weights=np.random.randn(input_dim,output_dim)*np.sqrt(2/input_dim)
        self.bias=np.zeros((1,output_dim))

    def forward_prop(self,input_data):
        self.input=input_data
        return np.dot(self.input,self.weights)+self.bias
    def backward_prop(self,output_grad,learning_rate):
        weight_grad=np.dot(self.input.T,output_grad)
        bias_grad=np.sum(output_grad,axis=0,keepdims=True)

        input_grad=np.dot(output_grad,self.weights.T)

        self.weights+=(learning_rate*weight_grad)
        self.bias+=(learning_rate*bias_grad)
        return input_grad
```

```python
class Sigmoid:
    def forward_prop(self,input_data):
        result=(1/(1+np.exp(-input_data)))
        self.output=result
        return result
    # why _ beside learning rate
    # In Python, if you have to accept a parameter but you don't plan on using it, we often rename it to _ or _lr. This tells other programmers (and the computer) that this variable is intentionally ignored.
    def backward_prop(self,output_grad,_learning_rate):
        # finding sensitivity
        return output_grad*(self.output*(1-self.output))
```

```python
class Relu:
    def forward_prop(self,input_data):
        return np.maximum(0,input_data)
    def backward_prop(self,output_grad,learning_rate):
        grad=output_grad.copy()
        grad[self.input<=0]=0
        return grad
```

```python
class NN:
    def __init__(self):
        # we use list here to access every layer in order pytorch has the same but under the hood the it's list but we can't see it while declaring
        self.layers=[
            Layer(2,6),
            Sigmoid(),
            Layer(6,1),
            Sigmoid()
        ]

    def predict(self,x):
        output=x
        for lay in self.layers:
            output=lay.forward_prop(output)
        return output
    def train(self,x,y,epochs=10000,learning_rate=0.1):
        for epoch in range(epochs):
            output=self.predict(x)
            gradient=y-output # if target-prediction we use += to update the weights and bias else if we use prediction-target we use -=

            for lay in reversed(self.layers):
                gradient=lay.backward_prop(gradient,learning_rate)
    
```
